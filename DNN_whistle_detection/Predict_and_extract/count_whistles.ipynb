```python
from count_whistles import howmanywhistlesinthisfile
import os
import csv

files_counts = {}
root_files = {}

for dirpath, dirnames, filenames in os.walk("/Users/apple/Desktop/Dolphins Patagonia/Analysis_2017/"):
    for file in filenames:
        if file.endswith(".csv"):
            csv_file = os.path.join(dirpath, file)
            file_name = os.path.basename(csv_file).split('.wav')[0]
            total_groups = howmanywhistlesinthisfile(csv_file)
            if total_groups != -1:
                files_counts[file_name] = total_groups
                root_file = os.path.basename(os.path.dirname(os.path.dirname(csv_file)))
                root_files[file_name] = root_file
                print(f"Total number of whistle groups in {file_name}: {total_groups}, Root file: {root_file}")

print(files_counts)

# Save the files_counts dictionary to a file
with open('whistles_count_2017.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Root File', 'File Name', 'Total Groups'])
    for key, value in files_counts.items():
        root_file = root_files[key]
        writer.writerow([root_file, key, value])
```
